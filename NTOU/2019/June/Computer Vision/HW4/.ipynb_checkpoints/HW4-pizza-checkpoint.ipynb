```python
import cv2
import numpy as np

totalFrame = 11
frame2 = cv2.imread('./dataset2/DSC_1158.JPG')
frameNameStart = 1158
frameIndexStart = 0
frameNum = 10
imgShrink = 16
frameEnlarge = 5


print('height:{} width:{}'.format(int(frame2.shape[0]),int(frame2.shape[1])))

def set_frame_number(x):
    global frame_num
    frame_num = x
    return

result = np.zeros((int(frame2.shape[0])//imgShrink*frameEnlarge,int(frame2.shape[1])//imgShrink*frameEnlarge,3))
count  = np.zeros((int(frame2.shape[0])//imgShrink*frameEnlarge,int(frame2.shape[1])//imgShrink*frameEnlarge))
ones = np.ones(((int(frame2.shape[0])//imgShrink,int(frame2.shape[1])//imgShrink)))

cv2.namedWindow('matching')

kpdetector = cv2.xfeatures2d.SIFT_create()
#kpdetector = cv2.AKAZE_create()

# create BFMatcher object
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)


while frameNum > -1:
    frame2 = cv2.imread('./dataset2/DSC_%d.JPG' % (frameNameStart + (frameNum + frameIndexStart) % totalFrame))
    
    frame2 = cv2.resize(frame2,(frame2.shape[1]//imgShrink,frame2.shape[0]//imgShrink))
    
#kp1, dt1 = kpdetector.detectAndCompute(frame1,None)
    gray = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
    kp2 = kpdetector.detect(gray,None)
    dt2 = kpdetector.compute(gray,kp2)[1]
    if frameNum == 10:
        T      = np.eye(3)
        T[0,2] = frame2.shape[1]
        T[1,2] = frame2.shape[0]
        result = cv2.warpPerspective(frame2,T,(result.shape[1],result.shape[0])).astype(np.float)
        t_count= cv2.warpPerspective(ones,T,(result.shape[1],result.shape[0])).astype(np.float)
        count += t_count.astype(np.float)
        disp = result.copy()
        cv2.imshow('stitched image',disp.astype(np.uint8))
    
        frame1 = frame2
        kp1 = kp2
        dt1 = dt2
    else:
        # Match descriptors.
        matches = bf.match(dt2,dt1)
        print('{}, # of matches:{}'.format(frameNum,len(matches)))

        # Sort in ascending order of distance.
        matches = sorted(matches, key = lambda x:x.distance)
        
        src = []
        dst = []
        for m in matches:
            src.append(kp2[m.queryIdx].pt + (1,))
            dst.append(kp1[m.trainIdx].pt + (1,))
            
        src = np.array(src,dtype=np.float)
        dst = np.array(dst,dtype=np.float)
    
        # find a homography to map src to dst
        A, mask = cv2.findHomography(src, dst, cv2.RANSAC) 
        
        # map to the first frame
        T = T.dot(A)
        warp_img = cv2.warpPerspective(frame2,T,(result.shape[1],result.shape[0])).astype(np.float)
        t_count  = cv2.warpPerspective(ones,T,(result.shape[1],result.shape[0])).astype(np.float)
        result+= warp_img
        count += t_count.astype(np.float)

        t_count= count.copy()
        t_count[t_count == 0] = 1
        disp = result.copy()
        
        disp[:,:,0] = result[:,:,0] / t_count
        disp[:,:,1] = result[:,:,1] / t_count
        disp[:,:,2] = result[:,:,2] / t_count
 
        cv2.imshow('stitched image',disp.astype(np.uint8))
   
        cv2.imshow('matching',cv2.drawMatches(frame2,kp2,frame1,kp1,matches[:1], None, flags=cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS))
        
        
        frame1 = frame2
        kp1 = kp2
        dt1 = dt2

    key = cv2.waitKey(20) & 0xFF
    if key == 27:
        break
    frameNum -= 1
cv2.waitKey()    
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

totalFrame = 12
frame2 = cv2.imread('./dataset2/DSC_1158.JPG')
frameNameStart = 1158
frameNum = 0
imgShrink = 16
frameEnlarge = 5


print('height:{} width:{}'.format(int(frame2.shape[0]),int(frame2.shape[1])))

def loadAllImage():
    images = np.zeros((totalFrame, int(frame2.shape[0])//imgShrink,int(frame2.shape[1])//imgShrink,3))
    for x in range(frameNum, totalFrame):
        frame = cv2.imread('./dataset2/DSC_%d.JPG' % (frameNameStart + (x+1) % totalFrame))
        frame = cv2.resize(frame,(frame.shape[1]//imgShrink,frame.shape[0]//imgShrink))
        images[x] = frame
    images = np.array(images,dtype=np.uint8)
    return images

def set_frame_number(x):
    global frame_num
    frame_num = x
    return

def phaseMatches(images, dt1, frameRemain):
    matchesDstMin = 1000000
    i = 0
    while i <= frameRemain:
        gray = cv2.cvtColor(images[i], cv2.COLOR_BGR2GRAY)
        kp2 = kpdetector.detect(gray,None)
        dt2 = kpdetector.compute(gray,kp2)[1]
        matches = bf.match(dt2,dt1)
        # Sort in ascending order of distance.
        matches = sorted(matches, key = lambda x:x.distance)
        
        j = 0
        matchesDstSum = 0
        for m in matches:
            if j > 50:
                break
            matchesDstSum += m.distance
            j+=1
            
        if(matchesDstSum < matchesDstMin) :
            matchesDstMin = matchesDstSum
            imgIndex = i
            maxMatches = matches
            dt2Max = dt2
            kp2Max = kp2
            imgMax = images[i].copy()
        i+=1
    return imgIndex, maxMatches, dt2Max, kp2Max, imgMax, matchesDstMin

def stitchingMatches(frameIndex, height, width):
    
    result = np.zeros((int(height)//imgShrink*frameEnlarge,int(width)//imgShrink*frameEnlarge,3))
    count  = np.zeros((int(height)//imgShrink*frameEnlarge,int(width)//imgShrink*frameEnlarge))
    ones = np.ones(((int(height)//imgShrink,int(width)//imgShrink)))
    
    sumMatchesDst = 0
    gray = cv2.cvtColor(images[frameIndex], cv2.COLOR_BGR2GRAY)
    kp2 = kpdetector.detect(gray,None)
    dt2 = kpdetector.compute(gray,kp2)[1]
    T      = np.eye(3)
    T[0,2] = images[frameIndex].shape[1]
    T[1,2] = images[frameIndex].shape[0]
    result = cv2.warpPerspective(images[frameIndex],T,(result.shape[1],result.shape[0])).astype(np.float)
    t_count= cv2.warpPerspective(ones,T,(result.shape[1],result.shape[0])).astype(np.float)
    count += t_count.astype(np.float)
    disp = result.copy()
    
    cv2.imshow('stitched image',disp.astype(np.uint8))
    key = cv2.waitKey(20) & 0xFF
    if key == 27:
        return 10e10, disp

    frame1 = images[frameIndex]
    kp1 = kp2
    dt1 = dt2
    
    tmpImages = images.copy()
    
    end = totalFrame - 1;
    
    tmpImages[frameIndex] = tmpImages[end]
    
    end -=1
    usedImgs = 1
    
    while end > -1:
        i, matches, dt2, kp2, frame2, matchDst = phaseMatches(tmpImages, dt1, end)
                
        usedImgs+=1
        
        sumMatchesDst += matchDst        
        
        src = []
        dst = []
        for m in matches:
            src.append(kp2[m.queryIdx].pt + (1,))
            dst.append(kp1[m.trainIdx].pt + (1,))
            
        src = np.array(src,dtype=np.float)
        dst = np.array(dst,dtype=np.float)
    
        # find a homography to map src to dst
        A, mask = cv2.findHomography(src, dst, cv2.RANSAC) 
        
        # map to the first frame
        T = T.dot(A)
        warp_img = cv2.warpPerspective(frame2,T,(result.shape[1],result.shape[0])).astype(np.float)
        t_count  = cv2.warpPerspective(ones,T,(result.shape[1],result.shape[0])).astype(np.float)
        result+= warp_img
        count += t_count.astype(np.float)

        t_count= count.copy()
        t_count[t_count == 0] = 1
        disp = result.copy()
        
        disp[:,:,0] = result[:,:,0] / t_count
        disp[:,:,1] = result[:,:,1] / t_count
        disp[:,:,2] = result[:,:,2] / t_count
        
 
        cv2.imshow('stitched image',disp.astype(np.uint8))
   
        cv2.imshow('matching',cv2.drawMatches(frame2,kp2,frame1,kp1,matches[:10], None, flags=cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS))
        tmpImages[i] = tmpImages[end]
        key = cv2.waitKey(20) & 0xFF
        if key == 27:
            break
        dt1 = dt2
        kp1 = kp2
        frame1 = frame2
        end-=1
    return sumMatchesDst/usedImgs, disp


cv2.namedWindow('matching')
cv2.namedWindow('best stitched image')
cv2.namedWindow('stitched image')

kpdetector = cv2.xfeatures2d.SIFT_create() 
#kpdetector = cv2.AKAZE_create()

# create BFMatcher object
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)

images = loadAllImage()
minStitch = 10e10

while frameNum < totalFrame:
    
    
    tmpMin, disp = stitchingMatches(frameNum, frame2.shape[0], frame2.shape[1])
    
    if tmpMin < minStitch:
        minStitch = tmpMin
        minDisp = disp
        
    cv2.imshow('best stitched image',minDisp.astype(np.uint8))
    
    key = cv2.waitKey(20) & 0xFF
    if key == 27:
        break
    frameNum += 1
cv2.waitKey()    
cv2.destroyAllWindows()
```
